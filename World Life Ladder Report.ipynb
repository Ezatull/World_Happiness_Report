# Introduction:

The World Happiness Report is a yearly publication that assesses and examines the happiness levels of populations across the globe. Beyond a mere ranking of nations by happiness, the report unfolds as an intricate narrative, delving into the reasons behind varying happiness levels worldwide. By envisioning happiness as a ladder, where the pinnacle signifies utmost joy and the base represents lesser contentment, the report allows us to comprehend the varying levels of happiness experienced in different regions.

However, this report's scope transcends mere measurement; it morphs into a metaphorical detective story, seeking to unveil the factors underlying divergent happiness levels. These factors encompass economic indicators, life expectancy, freedom to make choices, and the presence of social support networks. Furthermore, the report chronicles the dynamics of change over timeâ€”alterations that may either elevate or diminish happiness. Governments and organizations can leverage this information to enhance the quality of life within their countries.

Ultimately, the World Happiness Report is a profound exploration, not limited to statistical figures but deeply rooted in comprehending the human experience. # Exploratory Analysis & Visualization

Performing exploratory data analysis (EDA) on the 'data' DataFrame using the Seaborn and Matplotlib libraries.

```python
numeric_stats = data.describe()
numeric_stats
```

Descriptive Statistics:
Numeric statistics are calculated using the describe() function and stored in the numeric_stats variable.
These statistics provide information about the central tendency and spread of each numeric column in the DataFrame.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Variables',fontsize=16)
plt.show()
```

Correlation Heatmap:
The correlation matrix between numeric columns is calculated using the .corr() method on the numerical_columns.
A heatmap of the correlation matrix is created using Seaborn's heatmap() function.
The heatmap visualizes the relationships between numeric variables, with color intensity indicating the strength and direction of the correlation.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
plt.hist(data['Log GDP per capita'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Log GDP per capita')
plt.ylabel('Frequency')
plt.title('Distribution of Log GDP per capita')
plt.show()
```

Histogram:
A histogram is plotted using Matplotlib to visualize the distribution of the 'Log GDP per capita' column.
The x-axis represents the 'Log GDP per capita' values, and the y-axis represents the frequency of those values in the dataset.
The Log GDP per capita range from 0 to nearly 12 however, most of its values are around 9 to 11. 

```python
plt.figure(figsize=(12,10))
plt.scatter(data['Life Ladder'], data['Freedom to make life choices'], alpha=0.5)
plt.xlabel('Life Ladder')
plt.ylabel('Freedom to make life choices')
plt.title('Relationship between Life Ladder and Freedom to make life choices')
plt.show()
```

# Q & A

Average Life Ladder:

```python
Life_Ladder = data['Life Ladder'].mean()
Life_Ladder
```

The "Life Ladder" column represents a measure of life satisfaction or happiness. By calculating the mean of this column, you're essentially finding the average happiness score across all countries and years present in the dataset. The computed mean value of approximately 5.46 suggests that, on average, the sampled countries tend to report a moderate level of life satisfaction.

Country with the Highest GDP:

```python
highest_gdp_country = data.loc[data['Log GDP per capita'].idxmax(), 'Country name']
highest_gdp_country
```

The "Log GDP per capita" serves as an indicator of a country's economic prosperity. By identifying the country with the highest value in this column, you're pinpointing the nation that has the highest economic output per person. In this case, Luxembourg is recognized as having the highest per capita GDP, indicating its strong economic standing relative to other countries.

Distribution of Freedom Scores:

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Freedom to make life choices'], bins=15, color='green', alpha=0.7)
plt.xlabel('Freedom to make life choices')
plt.ylabel('Frequency')
plt.title('Distribution of Freedom to make life choices')
plt.show()
```

Visualizing the distribution of "Freedom to make life choices" scores through a histogram allows you to understand how this well-being factor is spread across the dataset. The histogram's x-axis showcases different ranges of freedom scores, while the y-axis displays the frequency of countries falling within those score ranges. This visualization range from 0.2 to nearly 1, however frequent values are between 0.7 to 0.9 meaning if a country hvae values within this interval it has similar freedom to most of the countries around the world.

Countires with lowest life expectancy:

```python
Lowest_life_expectancy=data.sort_values('Healthy life expectancy at birth',
                                        ascending=True).drop_duplicates(subset='Country name').head(5)
```

```python
Lowest_life_expectancy
```

Sorting the data by "Healthy life expectancy at birth" in ascending order allows you to identify the countries with the lowest values in terms of expected lifespan in good health. This subset of countries, including Haiti, Zimbabwe, Central African Republic, Lesotho, and Eswatini, presents a group facing significant challenges in terms of health and well-being.
